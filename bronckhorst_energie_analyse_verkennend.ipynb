## Context

Gemeente Bronckhorst werkt aan beleid rondom verduurzaming van de gebouwde omgeving,
zoals de warmtetransitie en het verminderen van energieverbruik.

In dit notebook wordt een **verkennende analyse** gegeven van:
- gemiddeld gas- en stroomverbruik per woning
- verschillen tussen buurten en wijken
- de relatie met woningkenmerken zoals WOZ-waarde en woningtype

De analyse is bedoeld als verkennend of oriënterend en kan verder nog worden uitgebreid en verdiept.

## Data

Voor deze analyse is gebruikgemaakt van openbare datasets, waaronder:

- Buurt- en wijkindelingen (CBS / gemeente)
- Gemiddeld energieverbruik per woning
- Woningkenmerken (o.a. WOZ-waarde, woningtype)
- Open geo-data voor kaartvisualisaties

De data is verwerkt via een eenvoudige **bronze → silver → gold**-structuur,
waarbij de gold-datasets analyse-klaar zijn gemaakt voor visualisatie in Power BI.

## Ruimtelijke verschillen in energieverbruik

Onderstaande kaarten tonen het gemiddeld gas- en stroomverbruik per woning
op buurt- en wijkniveau binnen Gemeente Bronckhorst.

![image_1](plots/1_gas_stroom_verbruik_buurt_wijk.png)

Observaties:
- Er zijn duidelijke ruimtelijke verschillen tussen buurten.
- Op wijkniveau zijn de verschillen kleiner, wat wijst op interne variatie binnen wijken.
- Gas- en stroomverbruik vertonen deels vergelijkbare patronen, maar niet volledig.

## Energieverbruik in relatie tot schaal en welvaart

Om beter te begrijpen waardoor verschillen ontstaan,
is gekeken naar de relatie tussen energieverbruik en:
- het aantal huishoudens per buurt
- de gemiddelde WOZ-waarde

![image_2](plots/2_gas_stroom_verbruik_hh_woz.png)

Observaties:
- Buurten met weinig huishoudens laten vaak een hoger gemiddeld gasverbruik zien.
- Er is een duidelijke positieve relatie tussen WOZ-waarde en energieverbruik,
  wat wijst op grotere woningen met hogere energievraag.

## Energieverbruik per woningtype

Tot slot is het gemiddeld gasverbruik uitgesplitst naar woningtype,
waarbij onder andere is gekeken naar:
- appartementen
- tussenwoningen
- hoekwoningen
- twee-onder-een-kap
- vrijstaande woningen

![image_3](plots/3_gas_verbruik_woz_woningtype.png)

Belangrijkste inzichten:
- Vrijstaande woningen hebben structureel het hoogste gasverbruik.
- Appartementen hebben het laagste gasverbruik.
- Ook bij vergelijkbare WOZ-waardes blijven verschillen tussen woningtypen zichtbaar.
- Woningtype is daarmee een belangrijke verklarende factor naast welvaart.

## Conclusie

Deze verkennende analyse laat zien dat:
- Energieverbruik binnen Gemeente Bronckhorst sterk varieert per buurt.
- WOZ-waarde en woningtype belangrijke verklarende factoren zijn.
- Met name vrijstaande woningen een relatief hoge energievraag hebben.

Deze inzichten kunnen worden gebruikt als eerste input voor
verdere beleidsanalyse rondom verduurzaming en warmtetransitie.
